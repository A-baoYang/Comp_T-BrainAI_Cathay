```python
app_id = '338922553215944'
app_secret = '6fba0ceda5334a6ec9b4d6bd56dffd5c'
access_token = 'EAAE0P45kC8gBABSrqJVMm1pZBjaBR4ZA6pdwycoQ7jlgzJVnLSGPzjDe92c7ZBZBnzYorLyoNC63YzxOiXlOgEaRblfcpITmpJaqfJjmMp6bYHJZBwhT5WhvlzdFVEIMkH6icJ02Lgk28N5r8AKBIIrlDsMsKY7No5mFgpZBY3EYYpiblwzWwaTXIDu7pfSjZCaWYJ8IrwNATGloZAG97lrs'
api = FacebookAdsApi.init(app_id, app_secret, access_token)

def interestSearchAll(query, filename):
    types = ['adinterest', 'adeducationschool', 'adeducationmajor', 'adworkemployer', 'adworkposition']
    ta_names = []
    
    for q in (query):
        print('---- ', q, ' ----')
        for t in types:
            print('>>>', t)

            params = {
                'q': q,
                'type': t,
                'limit': 100,
                'locale': 'zh_TW'
            }

            resp = TargetingSearch.search(
                params=params,
                api=api
            )


            for item in resp:
                trad_cn = item['name'].encode('unicode_escape').decode('unicode_escape')
                try:
                    if item['audience_size']:
                        print(trad_cn, ': ', item['audience_size'])
                        if q in trad_cn:
                            ta_names.append((trad_cn, item['audience_size']))

                    elif item['coverage']:
                        print(trad_cn, ': ', item['coverage'])
                        if q in trad_cn:
                            ta_names.append((trad_cn, item['coverage']))

                except:
                    pass


                try:
                    if item['disambiguation_category']:
                        print('disambiguation_category: ', item['disambiguation_category'].encode('unicode_escape').decode('unicode_escape'))
                except:
                    pass

                try:
                    if item['path']:
                        for index, path in enumerate(item['path']):
                            print('path_', index, ': ', path.encode('unicode_escape').decode('unicode_escape'))
                except:
                    pass
    
    
    with open(filename,'w') as out:
        csv_out=csv.writer(out)
        csv_out.writerow(['audience_name', 'size'])
        for row in ta_names:
            csv_out.writerow(row)
        
    return ta_names
```
