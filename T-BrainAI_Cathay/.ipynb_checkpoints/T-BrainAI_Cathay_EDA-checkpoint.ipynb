```python
# Y1=0抽樣抽 
# 'AGE':低/高, 'BMI': 0~0.1, 'CHARGE_CITY_CD': 避開A1/A2, 'EDUCATION_CD': nan/1/4, 'MARRIAGE_CD': nan/2, 'OCCUPATION_CLASS_CD': 0/5/6, 'IF_2ND_GEN_IND': Y, 'CUST_9_SEGMENTS_CD': C/H, 'IF_HOUSEHOLD_CLAIM_IND': N
# 'REBUY_TIMES_CNT': 'None','LEVEL': 2/3,'TERMINATION_RATE': 60~80,'BANK_NUMBER_CNT': 0.35~,'L1YR_LAPSE_IND': N:Y=1.5:1,'L1YR_PAYMENT_REMINDER_IND': N:Y=1.5:1
# 'LAST_A_CCONTACT_DT':N,'L1YR_A_ISSUE_CNT':3/4,'LAST_A_ISSUE_DT': N:Y=4:1,'L1YR_B_ISSUE_CNT': 1~,'LAST_B_ISSUE_DT':N:Y=1:1,'CHANNEL_A_POL_CNT':7~14,'CHANNEL_B_POL_CNT':2~,'LAST_B_CONTACT_DT':N:Y=2:1,'LAST_C_DT':N:Y=3:1,'L1YR_C_CNT':3~4
# 'APC_CNT':3~4,'APC_1ST_AGE':'None','APC_1ST_YEARDIF':0.5~0.75,'RFM_R':

for c in to_insur:
    if c not in nonNum_features:
        Y1_0[c] = Y1_0[c].fillna(-1)
        Y1_1[c] = Y1_1[c].fillna(-1)
        fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,8))
        ax1.scatter(Y1_0[c],Y1_0['CUS_ID'], label='Y1_0')
        ax1.set_title(c)
        ax2.scatter(Y1_1[c],Y1_1['CUS_ID'], label='Y1_1')
        ax2.set_title(c)
    if c in nonNum_features:
        Y1_0[c] = Y1_0[c].fillna('None')
        Y1_1[c] = Y1_1[c].fillna('None')
        fig,(ax3,ax4) = plt.subplots(1,2,figsize=(20,8))
        Y1_0.groupby(c)[c].count().plot(kind='bar',ax=ax3)
        Y1_1.groupby(c)[c].count().plot(kind='bar',ax=ax4)
        for tick in ax3.get_xticklabels():
            tick.set_fontproperties(font)
        for tick in ax4.get_xticklabels():
            tick.set_fontproperties(font)
        plt.show()
        
```

### 針對train資料集Y1=1的找特徵

```python
Y1_1 = df_train[df_train.index.isin(df_train_Y1.query('Y1 == [1]').index)]
Y1_1['index'] = Y1_1.index
Y1_0 = df_train[df_train.index.isin(df_train_Y1.query('Y1 == [0]').index)]
Y1_0['index'] = Y1_0.index

nonNum_features = []
for feature in Y1_1.columns:
    try:
        if Y1_1[feature].dtype == object:
            nonNum_features.append(feature)
    except TypeError as e:
        print(e)

print(f'{len(nonNum_features)} Non-num Features : {nonNum_features}\n')

Num_features = []
for feature in Y1_1.columns:
    try:
        if (Y1_1[feature].dtype == int) or (Y1_1[feature].dtype == float):
            Num_features.append(feature)
    except TypeError as e:
        print(e)

print(f'{len(Num_features)} Num Features : {Num_features}\n')
```

```python
name=[]
Y1_1_avg=[]
Y1_0_avg=[]
Y1_1_med=[]
Y1_0_med=[]

for item in Num_features:
    name.append(item)
    Y1_1_avg.append(Y1_1[item].mean())
    Y1_1_med.append(Y1_1[item].median())
    Y1_0_avg.append(Y1_0[item].mean())
    Y1_0_med.append(Y1_0[item].median())
t = pd.DataFrame(np.column_stack([name,Y1_1_avg,Y1_0_avg,Y1_1_med,Y1_0_med]), columns=['name','Y1_1_avg','Y1_0_avg','Y1_1_med','Y1_0_med'])
t

# for item in Num_features:
#     fig, axes = plt.subplots()
#     print(item, '- avg: ', Y1_1[item].mean())
#     Y1_1.boxplot(column=item, ax=axes)
    
#     axes.set_ylabel(f'values of {item}')
#     plt.show()
```
