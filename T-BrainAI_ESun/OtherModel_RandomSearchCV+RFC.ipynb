```python
import os, copy
import numpy as np
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
from datetime import date
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font = FontProperties(fname='font_ch.ttf', size=14)
import seaborn as sns

from sklearn.metrics import log_loss, roc_auc_score, auc, roc_curve, accuracy_score, f1_score, brier_score_loss, classification_report, confusion_matrix, make_scorer
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
# 因為擬合(fit)與編碼(transform)需要分開, 因此不使用.get_dummy, 而採用 sklearn 的 OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split, cross_val_score, RandomizedSearchCV
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
df_train = pd.read_csv('train.csv')
print(df_train.shape)
df_train


df_test = pd.read_csv('test.csv')
print(df_test.shape)
df_test
```

```python
## 缺值及unique數量檢查
print('df_train: ', len(df_train))
df_train.isnull().sum()
df_train = df_train.fillna('None')
for c in df_train.columns:
    print(c, ': ', np.unique(df_train[c].values), ' #', len(np.unique(df_train[c].values)))

print('df_test: ', len(df_test))
df_test.isnull().sum()
df_test = df_test.fillna('None')
for c in df_test.columns:
    print(c, ': ', np.unique(df_test[c].values), ' #', len(np.unique(df_test[c].values)))
```

```python
## feature分類
nonNum_features = []
for feature in df_train.columns:
    try:
        if df_train[feature].dtype == object:
            nonNum_features.append(feature)
    except TypeError as e:
        print(e)

print(f'{len(nonNum_features)} Non-num Features : {nonNum_features}\n')

Num_features = []
for feature in df_train.columns:
    try:
        if (df_train[feature].dtype == int) or (df_train[feature].dtype == float):
            Num_features.append(feature)
    except TypeError as e:
        print(e)

print(f'{len(Num_features)} Num Features : {Num_features}\n')
```

```python
## simple feature selection
df_train_backup = df_train.copy()
# df_train = df_train_backup.copy()
df_train = df_train[['txkey','acqic', 'csmcu','hcefg','etymd','scity','ecfg','mcc','iterm', 'insfg','stocn','flg_3dsmk','flbmk','fraud_ind']]

df_test_backup = df_test.copy()
# df_test = df_test_backup.copy()
df_test = df_test[['txkey','acqic', 'csmcu','hcefg','etymd','scity','ecfg','mcc','iterm', 'insfg','stocn','flg_3dsmk','flbmk']]
```

```python
## Label Encoding
# labelencoder = LabelEncoder()
# for i in nonNum_features:
#     try:
#         if (len(df_train[i])>0):
#             df_train[i] = labelencoder.fit_transform(df_train[i])
#     except:
#         print(i)

# for i in nonNum_features_t:
#     try:
#         if (len(df_test[i])>0):
#             df_test[i] = labelencoder.fit_transform(df_test[i])
#     except:
#         print(i)


## One hot encoding
df_train = pd.get_dummies(df_train)
df_train
df_test = pd.get_dummies(df_test)
df_test

# specific
# df_train = df_train.drop(['flg_3dsmk_None','flbmk_Y'],axis=1)
# df_train
# df_test = df_test.drop(['flg_3dsmk_None','flbmk_Y'],axis=1)
# df_test
```

```python
## dataset split
train_Y = df_train[['fraud_ind']]
df_train_copy = df_train.drop(['fraud_ind'],axis=1)
# df_test_copy = df_test.drop(['txkey'],axis=1)


train_X_values = df_train_copy.values
train_x, val_x, train_y, val_y = train_test_split(train_X_values, train_Y, test_size=0.45, random_state=560, shuffle=True)
train_x

test_x = df_test.values
test_x
```

```python
# RandomizedSearchCV - https://medium.com/@hjhuney/implementing-a-random-forest-classification-model-in-python-583891c99652

rf = RandomForestClassifier()

n_estimators = [int(x) for x in np.linspace(start=10, stop=60, num=10)]
min_samples_split = [int(x) for x in np.linspace(start=2, stop=20, num=10)]
min_samples_leaf = [int(x) for x in np.linspace(start=2, stop=200, num=10)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(3, 12, num=10)]
max_depth.append(None)

class_weight = [
    {0: 1., 1: 70.},
    {0: 1., 1: 60.},
    {0: 1., 1: 50.},
    {0: 1., 1: 40.},
    {0: 1., 1: 30.}
]

random_grid = {
    'n_estimators': n_estimators,
    'min_samples_split': min_samples_split,
    'min_samples_leaf': min_samples_leaf,
    'max_features': max_features,
    'max_depth': max_depth,
    'class_weight': class_weight
 }

rf_random = RandomizedSearchCV(estimator=rf, param_distributions=random_grid, n_iter=100, cv=3, verbose=2, random_state=42, n_jobs=-1)
rf_random.fit(train_x, train_y)
print(rf_random.best_params_)
```

```python
## use RandomizedSearchCV's best result 
rfc = RandomForestClassifier(n_estimators=15, min_samples_split=2, min_samples_leaf=2, max_depth=None, max_features='sqrt', class_weight={0: 1.0, 1:60.0})
rfc.fit(train_x, train_y)
rfc_predict = rfc.predict(val_x)
# rfc_predict_proba = rfc.predict_proba(val_x)
rfc_cv_score = cross_val_score(rfc, train_X_values, train_Y, cv=10, scoring='f1')
# for j in np.arange(0.4,0.6,0.01):
#     val_y_pred_binary = []
#     for i in rfc_predict_proba[:,1]:
#         val_y_pred_binary.append((1. if i > j else 0.))

# for i in np.arange(0.1, 0.3, 0.01):
#     val_x['predict_binary'] = np.where((val_x['predict_proba'] > i), 1., 0.)

#     f1 = f1_score(y_true=val_y.fraud_ind, y_pred=val_y_pred_binary)
#     print('F1 Score: {:.3f}'.format(f1), ' ', j)

print("=== Confusion Matrix ===")
print(confusion_matrix(val_y, rfc_predict))
print('\n')
print("=== Classification Report ===")
print(classification_report(val_y, rfc_predict))
print('\n')
print("=== All F1 Scores ===")
print(rfc_cv_score)
print('\n')
print("=== Mean F1 Score ===")
print("Mean AUC Score - Random Forest: ", rfc_cv_score.mean())
```
