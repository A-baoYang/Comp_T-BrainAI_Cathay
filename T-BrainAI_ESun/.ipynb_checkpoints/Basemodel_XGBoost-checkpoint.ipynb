```python
import os
import numpy as np
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
from datetime import date
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font = FontProperties(fname='font_ch.ttf', size=14)

import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss, roc_auc_score, auc, roc_curve, accuracy_score, f1_score, confusion_matrix, classification_report
# from sklearn.ensemble import RandomForestClassifier
```

```python
df_train = pd.read_csv('train.csv')
print(df_train.shape)
df_train


df_test = pd.read_csv('test.csv')
print(df_test.shape)
df_test
```

```python
## 缺值及unique數量檢查
print('df_train: ', len(df_train))
df_train.isnull().sum()
df_train = df_train.fillna('None')
for c in df_train.columns:
    print(c, ': ', np.unique(df_train[c].values), ' #', len(np.unique(df_train[c].values)))

print('df_test: ', len(df_test))
df_test.isnull().sum()
df_test = df_test.fillna('None')
for c in df_test.columns:
    print(c, ': ', np.unique(df_test[c].values), ' #', len(np.unique(df_test[c].values)))
```

```python
## 觀察資料分布
df_train.groupby('fraud_ind')['txkey'].count().reset_index()
```

```python
## feature分類
nonNum_features = []
for feature in df_train.columns:
    try:
        if df_train[feature].dtype == object:
            nonNum_features.append(feature)
    except TypeError as e:
        print(e)

print(f'{len(nonNum_features)} Non-num Features : {nonNum_features}\n')

Num_features = []
for feature in df_train.columns:
    try:
        if (df_train[feature].dtype == int) or (df_train[feature].dtype == float):
            Num_features.append(feature)
    except TypeError as e:
        print(e)

print(f'{len(Num_features)} Num Features : {Num_features}\n')
```

```python
## simple feature selection
df_train_backup = df_train.copy()
# df_train = df_train_backup.copy()
df_train = df_train[['txkey','acqic', 'csmcu','hcefg','etymd','scity','ecfg','mcc','iterm', 'insfg','stocn','flg_3dsmk','flbmk','fraud_ind']]

df_test_backup = df_test.copy()
# df_test = df_test_backup.copy()
df_test = df_test[['txkey','acqic', 'csmcu','hcefg','etymd','scity','ecfg','mcc','iterm', 'insfg','stocn','flg_3dsmk','flbmk']]
```

```python
## Label Encoding
# labelencoder = LabelEncoder()
# for i in nonNum_features:
#     try:
#         if (len(df_train[i])>0):
#             df_train[i] = labelencoder.fit_transform(df_train[i])
#     except:
#         print(i)

# for i in nonNum_features_t:
#     try:
#         if (len(df_test[i])>0):
#             df_test[i] = labelencoder.fit_transform(df_test[i])
#     except:
#         print(i)


## One hot encoding
df_train = pd.get_dummies(df_train)
# df_train
df_test = pd.get_dummies(df_test)
# df_test

# specific
df_train = df_train.drop(['flg_3dsmk_None','flbmk_Y'],axis=1)
df_train
df_test = df_test.drop(['flg_3dsmk_None','flbmk_Y'],axis=1)
df_test
```

```python
## corr map with categorical features

rs = np.random.RandomState(0)
corr = df_train.corr()#.sort_values('fraud_ind',ascending=False)
corr.style.background_gradient(cmap='coolwarm')
```

```python
## dataset split
df_train_result = df_train[['txkey','fraud_ind']]
df_train_copy = df_train.drop(['txkey','fraud_ind'],axis=1)
df_test_copy = df_test.drop(['txkey'],axis=1)
train_x, valid_x, train_y, valid_y = train_test_split(df_train_copy, df_train_result, test_size=0.45, random_state=33)
```

```python
## XGBoost
xgbc = XGBClassifier()
model_XGBC = xgbc.fit(train_x, train_y[['fraud_ind']])

valid_y_pred = xgbc.predict_proba(valid_x)
valid_x['predict_proba'] = valid_y_pred[:, 1]

# auc_score = roc_auc_score(y_true=valid_y.fraud_ind, y_score=valid_y_pred[:, 1])
# acc = accuracy_score(y_true=valid_y.fraud_ind, y_pred=valid_y_pred.argmax(axis=1))
# print('Validation AUC: {:.3f}, Accuracy: {:.3f}'.format(auc_score, acc))


## binaryzing y_pred
for i in np.arange(0.1, 0.3, 0.01):
    valid_x['predict_binary'] = np.where((valid_x['predict_proba'] > i), 1., 0.)

    f1 = f1_score(y_true=valid_y.fraud_ind, y_pred=valid_x[['predict_binary']])
    print('F1 Score: {:.3f}'.format(f1), i)


# valid_x['predict_binary'] = np.where((valid_x['predict_proba'] > 0.19), 1., 0.)

# print(classification_report(y_true=valid_y.fraud_ind, y_pred=valid_x[['predict_binary']]))
# print('total 1s: ', sum(valid_y.fraud_ind), ' / total 0s: ', len(valid_y)-sum(valid_y.fraud_ind))
# print(confusion_matrix(y_true=valid_y.fraud_ind, y_pred=valid_x[['predict_binary']]))
```

```python
print(df_test_copy.shape)

test_y_pred = xgbc.predict_proba(df_test_copy)
df_test_copy['predict_proba'] = test_y_pred[:, 1]
df_test_copy['predict_binary'] = np.where((df_test_copy['predict_proba'] > 0.19), 1., 0.)
print(df_test_copy.shape)

df_test_copy

output = pd.concat((df_test[['txkey']], df_test_copy['predict_binary']), axis=1)
print(output.shape)
output
```

```python
out = output.groupby('txkey', as_index=False).mean()
out = out[['txkey', 'predict_binary']]
out.columns = ['txkey', 'fraud_ind']
out.to_csv('output_191026_2_xgboost_allFeature_cate-oneHot_toBinary_threshold_0.19.csv')
out
```
